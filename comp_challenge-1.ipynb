Challenge: Rejection sampling is a useful method for generating random samples from a distribution f(x)that may be hard to sample from.
```python
# Import packages

import numpy as np
import random
from scipy.special import gamma, beta
import matplotlib.pyplot as plt 
from matplotlib import style
style.use("ggplot")
```

```python
# Set seed
seed = 98765
random.seed(seed)


# First we intialize variables and registers to be used

m = 100000      # set number of sample
n_accepted = 0  
n_tried = 0

f_dev = []      # register for storing accepted random deviates
g_dev = []      # register for storing candidate random deviates

some_alpha = 2.0    # choose some alpha
some_beta = 2.0     # choose some beta

a = 0       
b = 1
c_update = []

B=gamma(some_alpha)*gamma(some_beta)/gamma(some_alpha+some_beta)
c_caret = 1.000000001


while n_accepted < m:
    U = np.random.uniform(0,1)  # draw U ~ Unif(0,1)
    X = np.random.uniform(a,b)  # draw X ~ U(0,1)
    x_f = (X**(some_alpha-1) * (1-X**(some_beta-1)))/B  
    x_g = 1/(b-a)
    
    if U <= x_f/(x_g*c_caret):  
        f_dev.append(X)
        g_dev.append(X)
        n_accepted += 1
        n_tried += 1
    else:
        g_dev.append(X)
        n_tried += 1
        
    c_sup = max(c_caret,x_f/x_g)
    c_caret = c_sup   
    c_update.append(c_caret)
    p_accepted = n_accepted/n_tried
    
style.use('ggplot')

plt.title('Random Sampling from Beta Distribution Using Rejection Sampling')
plt.hist(g_dev, color = 'blue', label='candidate distribution --uniform(0,1)')
plt.hist(f_dev, color = 'red', label='target distribution --beta(2,2)')
plt.legend(loc = 'lower center')
plt.ylabel('Count')
plt.xlabel('X')

print('total samples to sample 100000 = ', n_tried)
print('acceptance rate =', p_accepted)
```
